# Learning from Abstract vs. Concrete Multimedia Texts


This notebook contains the script for modeling data from a research project on **multimedia learning from texts differing in abstractness**. In the study, participants (N = 64) read two multimedia texts (texts supplemented with pictures) with different levels of abstractness:

- **Concrete text**: about yoga asanas  
- **Abstract text**: about Nash equilibrium  

After reading, participants completed a knowledge assessment consisting of five tasks of increasing difficulty:  
1. Recognition  
2. Verification  
3. Transfer  
4. Concept definition  
5. Summary  

**Research question:**  
Do multimedia learning outcomes differ depending on text abstractness? Are these differences consistent across the various test tasks?

To capture both systematic effects and individual variability, we employ **linear mixed-effects models (LMMs)**.

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
from scipy.stats import chi2
```

## Helper Functions

These functions enable preprocessing, modeling, and visualization:
- `tidy_summary`: compact fixed-effect tables
- `fit_mixedlm`: subject-intercept MixedLM wrapper
- `to_category`: convert selected columns to categorical dtype
- `infer_text_from_item`: map item codes to abstract/concrete conditions
- `summarize_by_condition`: provide descriptive stats per text/condition
- `plot_violin_box`: color-blind-friendly violin + box overlays
- `_within_task_z`: helper for within-task z-scoring (used by the optional global model)


```python
COLORBLIND_BASE = {"abstract": "#EBC4E1", "concrete": "#7F517D"}
```

```python
def tidy_summary(model, decimals=3):
    """
    Create a compact summary dataframe of fixed effects:
    estimate, std_error, t_value, p_value, ci_lower, ci_upper
    """
    params = model.params
    se = model.bse
    tvals = getattr(model, "tvalues", params / se)
    pvals = getattr(model, "pvalues", pd.Series(np.nan, index=params.index))
    conf = model.conf_int()
    conf.columns = ["ci_lower", "ci_upper"]

    out = pd.concat(
        [
            params.rename("estimate"),
            se.rename("std_error"),
            tvals.rename("t_value"),
            pvals.rename("p_value"),
            conf,
        ],
        axis=1,
    )
    out = out.applymap(
        lambda x: round(x, decimals) if isinstance(x, (int, float, np.floating)) else x
    )

    return out


def fit_mixedlm(df, formula, group_col, reml=True):
    """Fit a MixedLM with a random intercept for group_col."""
    model = smf.mixedlm(
        formula,
        df,
        groups=df[group_col],
    ).fit(reml=reml)
    return model, tidy_summary(model)



def to_category(df, columns):
    """Convert selected columns to pandas 'category' dtype in-place."""
    for col in columns:
        if col in df.columns:
            df[col] = df[col].astype("category")
    return df


def infer_text_from_item(series):
    """Return 'abstract' if item code starts with 'A', else 'concrete'."""
    str_codes = series.astype(str)
    mask = str_codes.str.startswith("A", na=False)
    return pd.Series(np.where(mask, "abstract", "concrete"), index=series.index)


def summarize_by_condition(
    df,
    value_col="Accuracy",
    text_col="Text",
    extra_groups=None,
    decimals=3,
    dataset_name="",
):
    """Print count, mean, sd, and median per condition."""
    group_cols = []
    if extra_groups:
        group_cols.extend(extra_groups)
    group_cols.append(text_col)
    summary = (
        df.groupby(group_cols)[value_col]
        .agg(["count", "mean", "std", "median"])
        .rename(columns={"count": "n", "std": "sd"})
        .round(decimals)
    )
    title = dataset_name or "Dataset"
    print(f"\n{title} descriptive stats ({text_col}):")
    print(summary)


def plot_violin_box(
    df,
    value_col="Accuracy",
    text_col="Text",
    dataset_name="",
    hue=None,
    palette=None,
):
    """Draw a violin plot with an overlaid boxplot for abstract vs. concrete scores."""
    if palette is None:
        if hue:
            hue_levels = pd.unique(df[hue].dropna())
            colors = sns.color_palette("colorblind", len(hue_levels))
            palette = dict(zip(hue_levels, colors))
        else:
            palette = COLORBLIND_BASE
    dodge = bool(hue)
    plt.figure(figsize=(6, 4))
    sns.violinplot(
        data=df,
        x=text_col,
        y=value_col,
        hue=hue,
        dodge=dodge,
        scale="width",
        cut=0,
        inner=None,
        palette=palette,
    )
    sns.boxplot(
        data=df,
        x=text_col,
        y=value_col,
        hue=hue,
        dodge=dodge,
        width=0.2,
        showcaps=True,
        boxprops={"facecolor": "white", "zorder": 3},
        showfliers=False,
        palette=palette,
    )
    ax = plt.gca()
    ax.set_title(dataset_name or "Distribution by condition")
    ax.set_xlabel("Text type")
    ax.set_ylabel(value_col)
    handles, labels = ax.get_legend_handles_labels()
    if hue:
        unique_vals = pd.unique(df[hue].dropna())
        legend_entries = len(unique_vals)
        ax.legend(handles[:legend_entries], labels[:legend_entries], title=hue)
    else:
        legend = ax.get_legend()
        if legend:
            legend.remove()
    plt.tight_layout()
    plt.show()
```

## Data Preprocessing

### Recognition

```python
rec = pd.read_excel("Data/1.1_Recognition.xlsx")

rec = rec[["Subject", "Item", "Picture_Accuracy", "Text_Accuracy"]].copy()
rec["Text"] = infer_text_from_item(rec["Item"])
rec = to_category(rec, ["Subject", "Text"])

rec_text_task = (
    rec.groupby(["Subject", "Text"], as_index=False)["Text_Accuracy"].sum()
    .rename(columns={"Text_Accuracy": "Accuracy"})
)
rec_text_task["task_type"] = "Text Recognition"
rec_text_task = to_category(rec_text_task, ["Subject", "Text", "task_type"])

rec_picture_task = (
    rec.groupby(["Subject", "Text"], as_index=False)["Picture_Accuracy"].sum()
    .rename(columns={"Picture_Accuracy": "Accuracy"})
)
rec_picture_task["task_type"] = "Picture Recognition"
rec_picture_task = to_category(rec_picture_task, ["Subject", "Text", "task_type"])

summarize_by_condition(rec_text_task, dataset_name="Text Recognition")
plot_violin_box(rec_text_task, dataset_name="Text Recognition accuracy")

summarize_by_condition(rec_picture_task, dataset_name="Picture Recognition")
plot_violin_box(rec_picture_task, dataset_name="Picture Recognition accuracy")
```

### Verification

```python
ver = pd.read_excel("Data/1.2_Verification.xlsx")
ver = ver[["Subject", "Item", "Accuracy"]].copy()
ver["Text"] = infer_text_from_item(ver["Item"])
ver = to_category(ver, ["Subject", "Text"])

ver_task = (
    ver.groupby(["Subject", "Text"], as_index=False)["Accuracy"].sum()
    .rename(columns={"Accuracy": "Accuracy"})
)
ver_task["task_type"] = "Verification"
ver_task = to_category(ver_task, ["Subject", "Text", "task_type"])

summarize_by_condition(ver_task, dataset_name="Verification")
plot_violin_box(ver_task, dataset_name="Verification accuracy")
```

### Transfer

```python
trans = pd.read_excel("Data/1.3_Transfer.xlsx")
value_cols = [c for c in trans.columns if c != "Subject"]
trans_long = (
    trans.melt(id_vars=["Subject"], value_vars=value_cols, var_name="Item", value_name="Accuracy")
    .dropna(subset=["Accuracy"])
    .copy()
)
trans_long["Text"] = infer_text_from_item(trans_long["Item"])
trans_long = to_category(trans_long, ["Subject", "Text"])

trans_task = (
    trans_long.groupby(["Subject", "Text"], as_index=False)["Accuracy"].sum()
    .rename(columns={"Accuracy": "Accuracy"})
)
trans_task["task_type"] = "Transfer"
trans_task = to_category(trans_task, ["Subject", "Text", "task_type"])

summarize_by_condition(trans_task, dataset_name="Transfer")
plot_violin_box(trans_task, dataset_name="Transfer accuracy")
```

### Concept Definition

```python
defsum = pd.read_excel("Data/1.4-1.5_Definition_and_Summary.xlsx")

definition = (
    defsum[["Subject", "Text", "Definition"]]
    .rename(columns={"Definition": "Accuracy"})
    .copy()
)
definition = to_category(definition, ["Subject", "Text"])

summarize_by_condition(definition, dataset_name="Concept Definition")
plot_violin_box(definition, dataset_name="Concept definition accuracy")
```

### Summary

```python
summary_df = (
    defsum[["Subject", "Text", "Summary"]]
    .rename(columns={"Summary": "Accuracy"})
    .copy()
)
summary_df = to_category(summary_df, ["Subject", "Text"])

summarize_by_condition(summary_df, dataset_name="Summary")
plot_violin_box(summary_df, dataset_name="Summary accuracy")
```

## Dataframe Binding

```python
rec_text_task = rec_text_task.copy()
rec_picture_task = rec_picture_task.copy()
ver_task = ver_task.copy()
trans_task = trans_task.copy()
definition_task = definition.copy()
definition_task["task_type"] = "Concept Definition"
definition_task = to_category(definition_task, ["Subject", "Text", "task_type"])
summary_task = summary_df.copy()
summary_task["task_type"] = "Summary"
summary_task = to_category(summary_task, ["Subject", "Text", "task_type"])

combined = pd.concat(
    [
        rec_text_task,
        rec_picture_task,
        ver_task,
        trans_task,
        definition_task,
        summary_task,
    ],
    axis=0,
    ignore_index=True,
)

combined = to_category(combined, ["Subject", "Text", "task_type"])
```

## Test-Specific Linear Mixed Effect Models

```python
test_specs = [
    {"name": "Picture Recognition", "df": rec_picture_task},
    {"name": "Text Recognition", "df": rec_text_task},
    {"name": "Verification", "df": ver_task},
    {"name": "Transfer", "df": trans_task},
    {"name": "Concept Definition", "df": definition_task},
    {"name": "Summary", "df": summary_task},
]

per_test_results = {}

for spec in test_specs:
    name = spec["name"]
    d = spec["df"].dropna(subset=["Accuracy", "Text", "Subject"])
    meta_info = {
        "n_obs": len(d),
        "n_subjects": d["Subject"].nunique(),
    }
    if (len(d) < 5) or (meta_info["n_subjects"] < 2) or (d["Text"].nunique() < 2):
        per_test_results[name] = {"model": None, "summary": None, "note": "Insufficient data", "meta": meta_info}
        continue
    formula = "Accuracy ~ C(Text)"
    model, tidy = fit_mixedlm(d, formula, group_col="Subject", reml=True)
    per_test_results[name] = {"model": model, "summary": tidy, "meta": meta_info}
```

```python
print("Per-task LMM: Score ~ C(Text) + (1 | Subject)")

for spec in test_specs:
    name = spec["name"]
    res = per_test_results.get(name, {})
    print(f"\n--- {name} ---")
    if res.get("summary") is None:
        print(res.get("note", "No summary available"))
    else:
        print(res["summary"])
```

### Interpretation: Test-Specific Mixed Models
- `Picture Recognition`: Participants significantly better recognized pictures embedded in concrete texts (â‰ˆ+0.13 accuracy units, p<.001).
- `Text Recognition`: Accuracy did not differ between abstract and concrete texts (estimate near zero, p>.60), showing that text recognition is insensitive to abstractness.
- `Verification` & `Transfer`: Concrete materials showed moderate advantages (â‰ˆ+0.13 and +0.08, both p<.01), showing that participants better operated with the acquired knowledge and applied it to novel tasks while working with concrete texts.
- `Concept Definition`: Concrete texts showed better results (â‰ˆ+0.09, p<.01).
- `Summary`: Participants produced better summaries after reading concrete texts (â‰ˆ+0.40 on the 0â€“6 scale, pâ‰ˆ.03).

Overall, five of six tasks were solved more efficially after reading concrete texts compared to abstract ones; the absence of significant differences in recognizing text fragments (text recognition task) may indicate that despite participants remembered what has been written, they didn't use this knowledge in more complex setting.


## Global Mixed Model (z-scored accuracy)
This section pools all tasks, z-standardizes accuracy within task, and fits subject-intercept mixed models with and without a `Text Ã— task_type` interaction.


```python
def _within_task_z(series):
    std = series.std(ddof=0)
    if (std is None) or np.isclose(std, 0):
        return series * 0
    return (series - series.mean()) / std

# Standardize accuracy within each task to equalize scales
global_scaled = combined.dropna(subset=["Accuracy", "Text", "task_type", "Subject"]).copy()
global_scaled["Accuracy_z"] = global_scaled.groupby("task_type")["Accuracy"].transform(_within_task_z)
global_scaled = global_scaled.dropna(subset=["Accuracy_z"])

main_formula = "Accuracy_z ~ C(Text) + C(task_type)"
inter_formula = "Accuracy_z ~ C(Text) * C(task_type)"

global_main, global_main_tidy = fit_mixedlm(global_scaled, main_formula, group_col="Subject", reml=True)
global_inter, global_inter_tidy = fit_mixedlm(global_scaled, inter_formula, group_col="Subject", reml=True)

print("\n=== Global mixed-effects model (main effects) ===")
print(global_main_tidy)
print("\n=== Global mixed-effects model (with Text Ã— task interaction) ===")
print(global_inter_tidy)

def lr_test(model_small, model_big):
    lr_stat = 2 * (model_big.llf - model_small.llf)
    df_diff = model_big.df_modelwc - model_small.df_modelwc
    p_val = chi2.sf(lr_stat, df_diff)
    return {"LR stat": lr_stat, "df": df_diff, "p-value": p_val}

print("\nInteraction vs. main-effects LRT:")
print(lr_test(global_main, global_inter))

# Task-specific abstractness contrasts derived from the interaction model
fe_index = global_inter.fe_params.index
contrast_rows = []
base_effect = "C(Text)[T.concrete]"

for task in global_scaled["task_type"].cat.categories:
    contrast = np.zeros((1, len(fe_index)))
    if base_effect in fe_index:
        contrast[0, fe_index.get_loc(base_effect)] = 1
    interaction_name = f"C(Text)[T.concrete]:C(task_type)[T.{task}]"
    if interaction_name in fe_index:
        contrast[0, fe_index.get_loc(interaction_name)] = 1
    test_res = global_inter.t_test(contrast)
    contrast_rows.append(
        {
            "task_type": task,
            "estimate": float(test_res.effect),
            "std_error": float(test_res.sd),
            "t_value": float(np.squeeze(test_res.tvalue)),
            "p_value": float(np.squeeze(test_res.pvalue)),
        }
    )

contrast_df = pd.DataFrame(contrast_rows)
print("\nPer-task concrete minus abstract effect (from global model):")
print(contrast_df.round({"estimate": 3, "std_error": 3, "t_value": 3, "p_value": 3}))

```

### Interpretation: Global Mixed Model
- The main-effects fit reports a positive `C(Text)[T.concrete]` coefficient, confirming that (after z-scoring within task) concrete materials lead to higher scores overall even when task differences are controlled.
- The likelihood-ratio test comparing the interaction model to the main-effects model is significant, indicating that the abstractness effect is not uniform across tasks; the divergence is driven mostly by the flat effect in Text Recognition versus consistent concrete advantages elsewhere.
- Interaction coefficients and the rounded contrast table show which tasks deviate from the baseline: Picture Recognition, Verification, Transfer, Concept Definition, and Summary all retain positive concreteâ€“abstract gaps on the standardized scale, while Text Recognition remains near zero.
- Because the contrasts are expressed in z-units, their magnitudes are directly comparable. Larger positive estimates (e.g., Summary) flag tasks that benefit most from concrete wording, whereas estimates near zero highlight tasks insensitive to abstractness.
